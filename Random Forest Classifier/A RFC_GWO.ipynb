```python
import numpy as np
import random
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Define GWO function
def GWO(objf, lb, ub, dim, SearchAgents_no, Max_iter):
    # Initialize alpha, beta, and delta positions and scores
    Alpha_pos = np.zeros(dim)
    Alpha_score = float("inf")
    
    Beta_pos = np.zeros(dim)
    Beta_score = float("inf")
    
    Delta_pos = np.zeros(dim)
    Delta_score = float("inf")

    # Initialize positions of search agents
    Positions = np.random.uniform(0, 1, (SearchAgents_no, dim)) * (np.array(ub) - np.array(lb)) + np.array(lb)
    
    Convergence_curve = np.zeros(Max_iter)

    print("GWO is optimizing \"" + objf.__name__ + "\"")    

    # Main loop
    for l in range(Max_iter):
        for i in range(SearchAgents_no):
            # Return the search agents that go beyond the boundaries of the search space
            for j in range(dim):
                Positions[i, j] = round(np.clip(Positions[i, j], lb[j], ub[j]), 2)
            print(Positions)
            # Calculate objective function for each search agent
            fitness = objf(Positions[i, :])
            print(fitness)
            # Update Alpha, Beta, and Delta
            if fitness < Alpha_score:
                Alpha_score = fitness
                Alpha_pos = Positions[i, :].copy()
            
            if fitness > Alpha_score and fitness < Beta_score:
                Beta_score = fitness
                Beta_pos = Positions[i, :].copy()
            
            if fitness > Alpha_score and fitness > Beta_score and fitness < Delta_score: 
                Delta_score = fitness
                Delta_pos = Positions[i, :].copy()
        
        a = 2 - l * ((2) / Max_iter)

        # Update the position of search agents including omegas
        for i in range(SearchAgents_no):
            for j in range(dim):     
                r1 = random.random()
                r2 = random.random()
               
                A1 = 2 * a * r1 - a
                C1 = 2 * r2
                
                D_alpha = abs(C1 * Alpha_pos[j] - Positions[i, j])
                X1 = Alpha_pos[j] - A1 * D_alpha
                           
                r1 = random.random()
                r2 = random.random()
                
                A2 = 2 * a * r1 - a
                C2 = 2 * r2
                
                D_beta = abs(C2 * Beta_pos[j] - Positions[i, j])
                X2 = Beta_pos[j] - A2 * D_beta       
                
                r1 = random.random()
                r2 = random.random() 
                
                A3 = 2 * a * r1 - a
                C3 = 2 * r2
                
                D_delta = abs(C3 * Delta_pos[j] - Positions[i, j])
                X3 = Delta_pos[j] - A3 * D_delta    

                # Round positions to two decimal places
                X1 = round(X1, 2)
                X2 = round(X2, 2)
                X3 = round(X3, 2)
                
                print(X1 , X2 , X3 )
                Positions[i, j] = (X1 + X2 + X3) / 3
        
        Convergence_curve[l] = Alpha_score
    
    print("Alpha position =", Alpha_pos)
    print("Beta position =", Beta_pos)
    print("Delta position =", Delta_pos)

    return Alpha_pos, Alpha_score
```

```python
# Define the objective function
def objf(X):
    # Extract hyperparameters from X
    n_estimators = int(round(X[0]))
    max_depth = int(round(X[1]))
    min_samples_split = int(round(X[2]))
    min_samples_leaf = int(round(X[3]))
    
    # Initialize RFC with specified hyperparameters
    rfc = RandomForestClassifier(n_estimators=n_estimators,
                                 max_depth=max_depth,
                                 min_samples_split=min_samples_split,
                                 min_samples_leaf=min_samples_leaf,
                                 random_state=42)
    # Train GBC model
    rfc.fit(X_train_sc, y_train)
    
    # Predict using GBC model
    y_pred = rfc.predict(X_test_sc)
    
    # Compute accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    
    # Return negative accuracy score for minimization
    return -accuracy
    
# Example usage
# Define the search space (hyperparameter ranges)
# Assuming each hyperparameter range is defined as a tuple (min_value, max_value)
n_estimators_range = (50, 200)
max_depth_range = (10, 50)
min_samples_split_range = (2, 20)
min_samples_leaf_range = (1, 10)

# Example GWO hyperparameters
dim = 4  # Dimensionality of the search space
num_wolves = 5  # Number of search agents
num_iterations = 100  # Maximum number of iterations

# Initialize lower bounds and upper bounds for hyperparameters
lb = [n_estimators_range[0], max_depth_range[0], min_samples_split_range[0], min_samples_leaf_range[0]]
ub = [n_estimators_range[1], max_depth_range[1], min_samples_split_range[1], min_samples_leaf_range[1]]

best_params, best_score = GWO(objf, lb, ub, dim, num_wolves, num_iterations)
print("Best parameters:", best_params)
print("Best score:", best_score)
```
